# Buat ekstraksi PDF dan simpan hasil vector embedding nya ke vector database dengan txtai dengan sentence tranformer `sentence-transformers/all-MiniLM-L6-v2`

```python
# Inisialisasi text splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

# Inisialisasi vector store txtai
index = Embeddings({
    "method": "transformers",
    "path": "sentence-transformers/all-MiniLM-L6-v2"
})

# Path ke direktori PDF
pdf_directory = "/content/all_pdfs"
data = []
doc_id = 0  # counter ID dokumen unik

# Proses semua PDF dalam direktori
for filename in os.listdir(pdf_directory):
    if filename.endswith(".pdf"):
        pdf_path = os.path.join(pdf_directory, filename)
        print(f"📄 Memproses: {filename}")

        # Load dan split
        loader = PyPDFLoader(pdf_path)
        documents = loader.load()
        chunks = text_splitter.split_documents(documents)

        # Format dan tambahkan ke dataset
        for chunk in chunks:
            text = chunk.page_content.replace("\n", " ").strip()
            metadata = {
                "source": filename,
                "page": chunk.metadata.get("page", None)
            }
            data.append((str(doc_id), text, metadata))
            doc_id += 1

# Indexing
print(f"\n🔍 Mulai indexing {len(data)} chunks...")
start_time = time.time()
index.index(data)
index.save("pdf_index")
end_time = time.time()
print(f"✅ Indexing selesai dalam {end_time - start_time:.2f} detik")
```

```python
import json

# export data to json
with open("pdf_index.json", "w") as f:
    json.dump(data, f)
```

```python
import json
from txtai.embeddings import Embeddings
import google.generativeai as genai
import os

GOOGLE_API_KEY="YOUR_GEMINI_API_KEY" #@param {type:"string"}

# Set API Key Gemini
genai.configure(api_key=GOOGLE_API_KEY)

# Load data JSON (list of tuples: [id, text, metadata])
with open("pdf_index.json", "r") as f:
    data = json.load(f)

# Load txtai index
index = Embeddings()
index.load("pdf_index")

# Inisialisasi model Gemini (native)
model = genai.GenerativeModel("gemini-2.0-flash")  # atau "gemini-pro" / "gemini-1.5-pro"

# Fungsi RAG
def ask_gemini_rag(question: str):
    results = index.search(question, 3)
    print("🔍 Results:", results)

    context_chunks = []
    metadata_sources = []

    for result in results:
        doc_id = str(result[0])
        print(f"🔎 Mencari doc_id: {doc_id}")

        found = False
        for i in range(len(data)):
            if str(data[i][0]) == doc_id:
                text = data[i][1]
                metadata = data[i][2]
                context_chunks.append(text)
                metadata_sources.append(metadata)
                print(f"✅ Ditemukan: {metadata}")
                found = True
                break

        if not found:
            print(f"⚠️ Dokumen dengan ID {doc_id} tidak ditemukan di JSON")

    if not context_chunks:
        return "Tidak ada konteks yang ditemukan untuk menjawab."

    # Gabungkan konteks dan info sumber
    context = "\n\n".join(context_chunks)
    source_info = "\n".join(
        [f"- {meta['source']}, halaman {meta.get('page', '?')}" for meta in metadata_sources]
    )

    # Prompt untuk Gemini
    prompt = f"""
Berikut adalah potongan dokumen yang relevan:\n\n{context}

Pertanyaan:
{question}

Jawablah dengan lengkap dan akurat berdasarkan konteks di atas.
Cantumkan sumber file dan halaman jika relevan.

Sumber:
{source_info}
"""

    # Kirim ke Gemini
    response = model.generate_content(prompt)
    return response.text
```

```python
#@title Buat file kode app.py untuk program streamlit
%%writefile app.py
import os
import streamlit as st
import json
from txtai.embeddings import Embeddings
import google.generativeai as genai
import os

GOOGLE_API_KEY="YOUR_GEMINI_API_KEY" #@param {type:"string"}

# Set API Key Gemini
genai.configure(api_key=GOOGLE_API_KEY)

# Load data JSON (list of tuples: [id, text, metadata])
with open("pdf_index.json", "r") as f:
    data = json.load(f)

# Load txtai index
index = Embeddings()
index.load("pdf_index")

# Inisialisasi model Gemini (native)
model = genai.GenerativeModel("gemini-2.0-flash")  # atau "gemini-pro" / "gemini-1.5-pro"

# Fungsi RAG
def ask_gemini_rag(question: str):
    results = index.search(question, 3)
    print("🔍 Results:", results)

    context_chunks = []
    metadata_sources = []

    for result in results:
        doc_id = str(result[0])
        print(f"🔎 Mencari doc_id: {doc_id}")

        found = False
        for i in range(len(data)):
            if str(data[i][0]) == doc_id:
                text = data[i][1]
                metadata = data[i][2]
                context_chunks.append(text)
                metadata_sources.append(metadata)
                print(f"✅ Ditemukan: {metadata}")
                found = True
                break

        if not found:
            print(f"⚠️ Dokumen dengan ID {doc_id} tidak ditemukan di JSON")

    if not context_chunks:
        return "Tidak ada konteks yang ditemukan untuk menjawab."

    # Gabungkan konteks dan info sumber
    context = "\n\n".join(context_chunks)
    source_info = "\n".join(
        [f"- {meta['source']}, halaman {meta.get('page', '?')}" for meta in metadata_sources]
    )

    # Prompt untuk Gemini
    prompt = f"""
Berikut adalah potongan dokumen yang relevan:\n\n{context}

Pertanyaan:
{question}

Jawablah dengan lengkap dan akurat berdasarkan konteks di atas.
Cantumkan sumber file dan halaman jika relevan.

Sumber:
{source_info}
"""

    # Kirim ke Gemini
    response = model.generate_content(prompt)
    return response.text


# Streamlit UI
st.set_page_config(page_title="PDF RAG dengan Gemini", layout="wide")
st.title("📄 PDF Question Answering (RAG) dengan Gemini")

query = st.text_input("📝 Masukkan pertanyaan Anda:")
submit = st.button("🔍 Cari Jawaban")

if submit and query:
    with st.spinner("Memproses dengan Gemini..."):
        response = ask_gemini_rag(query)
    st.markdown("### 🧠 Jawaban:")
    st.write(response)
```
